```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris Dataset
iris_data = load_iris()

# Create a DataFrame from the dataset
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the target variable to the DataFrame
iris_df['target'] = iris_data.target

# Map numerical target values to class names
iris_df['species'] = iris_df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display the first few rows of the DataFrame
print(iris_df.head())

# Display summary statistics
print(iris_df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot for feature comparison
sns.pairplot(iris_df, hue='species')
plt.savefig('Pairplot for feature comparison.png')
plt.show()

# Boxplot for feature distributions
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.drop(columns=['target', 'species']))
plt.title('Feature Distributions')
plt.show()
```

```python
# Calculate correlation matrix excluding the 'species' column
corr_matrix = iris_df.drop(columns=['species']).corr()

# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Distribution plots for each feature by species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_data.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=iris_df, x=feature, hue='species', kde=True, element="step", stat="density", common_norm=False)
    plt.title(f'Distribution of {feature} by Species')
plt.tight_layout()
plt.show()
```

```python
# Violin plots for each feature by species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_data.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(data=iris_df, x='species', y=feature)
    plt.title(f'Violin Plot of {feature} by Species')
plt.tight_layout()
plt.show()
```

```python
# Swarm plots for each feature by species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_data.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.swarmplot(data=iris_df, x='species', y=feature)
    plt.title(f'Swarm Plot of {feature} by Species')
plt.tight_layout()
plt.show()
```

```python
# Pairwise scatter plot of features colored by species
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.title('Pairwise Scatter Plot of Features')
plt.show()
```
